# 연습하기

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    '상품': ['키보드', '마우스', '모니터'],
    '가격': [30000, 15000, 200000],
    '재고': [10, 25, 7]
})
df
```

```python
prod_s = df['상품']
prod_s
```

```python
prod_df = df[['상품']]
prod_df
```

```python
expensive = df.loc[df['가격'] >= 20000, ['상품', '가격']]
expensive
```

```python
df.loc[df['상품'] == '마우스', '가격'] = 17000
df
```

```python
sub_iloc = df.iloc[0:3, 0:2]
sub_iloc
```

```python
df['부가세포함가'] = (df['가격'] * 1.1).astype(int)
df
```

```python
df_sorted_desc = df.sort_values(by = "가격", ascending = False)
df_sorted_desc
```

```python
df = pd.DataFrame({
    "메뉴": ['돼지국밥', '순대국밥', '내장국밥', '섞어국밥', '뼈다귀해장국', '소머리국밥', '설렁탕'],
    "가격": [4500, 5000, 5500, 5000, 4000, 5900, 5300],
    "판매수량": [10, 110, 250, 110, 20, 210, 0],
})
df
```
